```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_set = pd.read_csv("Human_activity_Dataset.csv")
#setting the precision to 4 decimal places
pd.set_option("display.precision",4 )
#setting the display of maximum column to null - this will display all the rows in the dataset  
pd.set_option("display.max.columns", 10)
print("Number of rows in the dataset: ", len(data_set))
print("Dimension of the of the dataset: ", data_set.shape)
data_set.info()
```

```python
#checking the first 2 rows of data in the begining of the dataset
print("Data Check first two rows:\n" , data_set.head(2))
#checking the first 2 rows of data in the begining of the dataset
print("\nData Check last two rows:\n" , data_set.tail(2))
```

```python
print("Description\n", data_set.describe())
```

```python
# displaying basic statistics for the dataset
data_set.describe(include=object)
```

```python
#Activity count
print("Activity Count:")
activity_counts = data_set["Activity"].value_counts()
print(activity_counts)
plt.title("\n\nPie Chart - % Distribution of Activities ", fontsize=50)
activity_counts.plot(kind='pie',  autopct='%.0f%%', figsize =(10,15), startangle=90, fontsize=15)
plt.show()
print("Based on the Pie Chart above, the human activities seems to be equally distributed from the data collected\n")




```

```python
group = data_set[['subject/Participant','Activity']].value_counts().reset_index().groupby("subject/Participant")

fig, axs = plt.subplots(6,5,figsize=(25, 25))
for (count, grp), ax in zip(group, axs.flat):
    ax.pie(grp[0], labels=grp.Activity,autopct='%.0f%%',startangle=90)
    ax.set_title("Plot"+format(count))

        
plt.show()
print("From the graphs it seems participation for each activity is fairly distributed")
```
